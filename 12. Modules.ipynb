##  12.1  Changing Python's default import path <a name='Modules-Default-Import-Path'></a>

The Python interpreter supports two types of modules:
-  built-ins, which are compiled into Python
-  file-based modules, which are accessed via the interpreter's resident file system.

Python searches for file-based modules in the directories listed in `sys.path`.  It processes a statement like `import foo` by searching the directories named in `sys.path` from left to right, selecting the first "foo.py" in the list.  This is similar to how
-  the POSIX CLI searches directories named in the `PATH` environment variable for executables
-  Java's runtime engine searches directories named in `CLASSPATH` for Java class libraries.

Updating `sys.path` changes how Python searches for modules.

The following examples use these Python library resources:
-  `os.mkdir` - creates a directory
-  `sys.path` - names directories that Python will search for modules
-  `importlib.invalidate_caches` - clear Python's import cache, to assure subsequent, 'clean' executions of example
-  `shutil.rmtree` - removes a directory tree, including all files therein


```python
# 12.1.a  Accessing a module in a subdirectory of the current directory that's not in sys.path
# example should fail, since the module's directory isn't named in sys.path

# supporting constants
CREATE_AS_NEW_FILE = 'x'    # file access mode for open()

# library resources
import os, sys, importlib, shutil

# supporting functions
make_printable = lambda exception: '' if str(exception) is None else str(exception)

# precondition check: ensure that we're creating a totally new, unreferenced module in a totally new directory

sample_dir_name = 'sample_directory'
corrective_action = f'please change {sample_dir_name} to a directory name that is not in sys.path'
assert sample_dir_name not in sys.path, corrective_action

sample_module_name = 'sample_module'  
corrective_action = f'please change {sample_module_name} to a module name that is not in use'
assert sample_module_name not in sys.modules.keys(), corrective_action

sample_function_name = 'sample_function' 
corrective_action = f'please change {sample_function_name} to a function name that is not in use'
assert sample_function_name not in dir(), corrective_action

importlib.invalidate_caches()   # clear stale module data, possibly from previous exercise

# preconditions established: create a module that's not on sys.path, then try to import and use it

file_to_access = sample_dir_name + '/' + sample_module_name + '.py'

try:
  os.mkdir( sample_dir_name )
  try:
    function_header = f'def {sample_function_name}(x, y):'
    return_message  = f'{sample_module_name}.{sample_function_name} in {sample_dir_name}: x+y is {{x+y}}' 
    function_body   = "return f'" + return_message + "'" 
    file_to_access_fd = open( file_to_access, CREATE_AS_NEW_FILE )
    file_to_access_fd.write( function_header + '\n' )
    file_to_access_fd.write( '  ' + function_body + '\n' )
    file_to_access_fd.close()
    try:
      exec( f'import {sample_module_name}' )
      result_from_sample_function = eval( f'{sample_module_name}.{sample_function_name}(3, 4)' )
      print( result_from_sample_function )
    except Exception as exception:
      print( f"can't access {sample_module_name}: {make_printable(exception)}" )
  except Exception as exception:
    print( f"can't create file ({file_to_access}): {make_printable(exception)}" )
  finally:
    try:
      shutil.rmtree( sample_dir_name )
    except Exception as exception:
      print( f"can't remove directory ({sample_dir_name}): {make_printable(exception)}" )
except Exception as exception:
  print( f"can't create directory ({sample_dir_name}): {make_printable(exception)}" )
```

```python
# 12.1.b  Accessing a module in a subdirectory of the current directory that's not in sys.path
# previous example, repeated with appropriate import statement

# supporting constants
CREATE_AS_NEW_FILE = 'x'    # file access mode for open()

# library resources
import os, sys, importlib, shutil

# supporting functions
make_printable = lambda exception: '' if str(exception) is None else str(exception)

# precondition check: ensure that we're creating a totally new, unreferenced module in a totally new directory

sample_dir_name = 'sample_directory'
corrective_action = f'please change {sample_dir_name} to a directory name that is not in sys.path'
assert sample_dir_name not in sys.path, corrective_action

sample_module_name = 'sample_module'  
corrective_action = f'please change {sample_module_name} to a module name that is not in use'
assert sample_module_name not in sys.modules.keys(), corrective_action

sample_function_name = 'sample_function' 
corrective_action = f'please change {sample_function_name} to a function name that is not in use'
assert sample_function_name not in dir(), corrective_action

importlib.invalidate_caches()   # clear stale module data, possibly from previous exercise

# preconditions established: create a module that's not on sys.path, then try to import and use it

file_to_access = sample_dir_name + '/' + sample_module_name + '.py'

try:
  os.mkdir( sample_dir_name )
  try:
    function_header = f'def {sample_function_name}(x, y):' 
    return_message = f'{sample_module_name}.{sample_function_name} in {sample_dir_name}: x+y is {{x+y}}' 
    function_body  = "return f'" + return_message + "'" 
    file_to_access_fd = open( file_to_access, CREATE_AS_NEW_FILE )
    file_to_access_fd.write( function_header + '\n' )
    file_to_access_fd.write( '  ' + function_body + '\n' )
    file_to_access_fd.close()
    try:
      # update path before accessing the module 
      sys.path += [sample_dir_name]
      exec( f'import {sample_module_name}' )
      result_from_sample_function = eval( f'{sample_module_name}.{sample_function_name}(3, 4)' )
      print( result_from_sample_function )
    except Exception as exception:
      print( f"can't access {sample_module_name}: {make_printable(exception)}" )
    finally:
      # undo changes to the environment 
      sys.path.remove( sample_dir_name )
      del sys.modules[ sample_module_name ]
      importlib.invalidate_caches()
  except Exception as exception:
    print( f"can't create file ({file_to_access}): {make_printable(exception)}" )
  finally:
    try:
      shutil.rmtree( sample_dir_name )  # clean up the directory tree
    except Exception as exception:
      print( f"can't remove directory ({sample_dir_name}): {make_printable(exception)}" )
except Exception as exception:
  print( f"can't create directory ({sample_dir_name}): {make_printable(exception)}" )
```

##  12.2  Module caching and reimportation <a name='Modules-Caching-And-Reimportation'></a>

When Python first imports a module 'foo', it caches two of foo's attributes in `sys.modules`:
-  the name of the module being imported, in a key in `sys.modules`
-  the module's location, in the key's corresponding value

Subsequent attempts to reimport 'foo' then either
-  have no effect, if foo is present in the session, or
-  reference the module location recorded in sys.modules independently of `sys.path`, otherwise.

The import statement's caching of previously imported modules speeds the importation of new modules when those modules, in turn, reference previously imported modules.  Caching, however, also interferes with the updating of code over the course of a given session.

In order to reset an imported module dynamically,
-  first use `importlib.invalidate_caches` to invalidate the current session's cache
-  then use `importlib.reload` to reload the module.

```python
# 12.2  Showing effect of import path and cache management on module reloading
#
# supporting constants
CREATE_AS_NEW_FILE = 'x'    # file access mode for open()

# library resources
import os, sys, importlib, shutil

# supporting functions
make_printable = lambda exception: '' if str(exception) is None else str(exception)

# precondition check: ensure that we're creating a totally new, unreferenced module in a totally new directory

sample_dir_name_1 = 'sample_dir_1'
corrective_action = f'please change {sample_dir_name_1} to a directory name that is not in sys.path'
assert sample_dir_name_1 not in sys.path, corrective_action

sample_dir_name_2 = 'sample_dir_2'
corrective_action = f'please change {sample_dir_name_2} to a directory name that is not in sys.path'
assert sample_dir_name_2 not in sys.path, corrective_action

sample_module_name = 'sample_module' 
corrective_action = f'please change {sample_module_name} to a module name that is not in use'
assert sample_module_name not in sys.modules.keys(), corrective_action

sample_function_name = 'sample_function' 
corrective_action = f'please change {sample_function_name} to a function name that is not in use'
assert sample_function_name not in dir(), corrective_action

# preconditions established: create a module that's not on sys.path, then try to import it

file_to_import_1 = sample_dir_name_1 + '/' + sample_module_name + '.py'
file_to_import_2 = sample_dir_name_2 + '/' + sample_module_name + '.py'

try:
  os.mkdir( sample_dir_name_1 )
  try:
    function_header = f'def {sample_function_name}(x, y):'
    return_message  = f'{sample_module_name}.{sample_function_name} in {sample_dir_name_1}: x+y is {{x+y}}' 
    function_body   = "return f'" + return_message + "'" 
    file_to_import_1_fd = open( file_to_import_1, CREATE_AS_NEW_FILE )
    file_to_import_1_fd.write( function_header + '\n' )
    file_to_import_1_fd.write( '  ' + function_body + '\n' )
    file_to_import_1_fd.close()
    try:
      os.mkdir( sample_dir_name_2 )
      try:
        function_header = f'def {sample_function_name}(x, y):'
        return_message  = f'{sample_module_name}.{sample_function_name} in {sample_dir_name_2}: x-y is {{x-y}}' 
        function_body   = "return f'" + return_message + "'" 
        file_to_import_2_fd = open( file_to_import_2, CREATE_AS_NEW_FILE ) 
        file_to_import_2_fd.write( function_header + '\n' )
        file_to_import_2_fd.write( '  ' + function_body + '\n' )
        file_to_import_2_fd.close()
        try:
          # give precedence to sample_dir_name_1, then import and access the module 
          # 
          sys.path += [sample_dir_name_1, sample_dir_name_2] 
          print( '>> importing', sample_module_name, '-', sample_dir_name_1, 'precedes', sample_dir_name_2, 'in path' )
          exec( f'import {sample_module_name}' )
          result_from_sample_function = eval( f'{sample_module_name}.{sample_function_name}(3, 4)' )
          print( result_from_sample_function, '\n' )
          # 
          # give precedence to sample_dir_name_2, then import and access the module 
          #
          print( '>> reversing', sample_dir_name_1, 'and', sample_dir_name_2, 'in path' )
          sys.path = sys.path[:len(sys.path)-2] + [sample_dir_name_2, sample_dir_name_1] 
          result_from_sample_function = eval( f'{sample_module_name}.{sample_function_name}(3, 4)' )
          print( result_from_sample_function, '\n' )
          # 
          # retry previous, but resetting and reloading Python's import cache first
          #
          print( '>> invalidating cache and reimporting', sample_module_name )
          importlib.invalidate_caches() 
          exec( f'importlib.reload( {sample_module_name} )' ) 
          result_from_sample_function = eval( f'{sample_module_name}.{sample_function_name}(3, 4)' )
          print( result_from_sample_function )
        except Exception as exception:
          print( f"can't access {sample_module_name}: {make_printable(exception)}" )
        finally:
          # undo changes to the environment 
          sys.path.remove( sample_dir_name_1 )
          sys.path.remove( sample_dir_name_2 )
          del sys.modules[ sample_module_name ]
          importlib.invalidate_caches()
      except Exception as exception:
        print( f"can't create file ({file_to_import_2}): {make_printable(exception)}" )
      finally:
        try:
          shutil.rmtree( sample_dir_name_2 )
        except Exception as exception:
          print( f"can't remove directory ({sample_dir_name_2}): {make_printable(exception)}" )
    except Exception as exception:
      print( f"can't create directory ({sample_dir_name_2}): {make_printable(exception)}" )
  except Exception as exception:
    print( f"can't create file ({file_to_import_1}): {make_printable(exception)}" )
  finally:
    try:
      shutil.rmtree( sample_dir_name_1 )  # clean up the first directory tree
    except Exception as exception:
      print( f"can't remove directory ({sample_dir_name_1}): {make_printable(exception)}" )
except Exception as exception:
  print( f"can't create directory ({sample_dir_name_1}): {make_printable(exception)}" )
```
